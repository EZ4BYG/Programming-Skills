```python
import numpy as np
from numpy import random  
import matplotlib.pyplot as plt
from IPython.display import clear_output, display   # needed only in Jupyter to plot 2D matrix dramatically, no need in IDE
```

```python
A = np.zeros((50,100))
```

```python
# Display
fig, ax = plt.subplots(figsize = (12,6))
t = 0
while t < 10:
    A = A + random.rand(50,100)
    # Plotting
    image = ax.imshow(A)
    # Display
    clear_output(wait = True)  # 'wait = True' is essential!
    display(fig)               # 'display' is essential!
    # Update the time
    t = t + 1
```
